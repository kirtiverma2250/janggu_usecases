# Downloads for DeepSEA and DanQ binding prediction

This notebook downloads data required for the DeepSEA and DanQ use case.

To run the notebook, bedtools, samtools and wget are required

```python
import os
import pandas as pd
import urllib.request
```

```python
output = '../02_deepsea_danq_prediction/data'
os.makedirs(output, exist_ok=True)
```

```python
df = pd.read_csv('../extra/urls.txt', header=None)
```

```python
for idx, row in df.iterrows():
    print(os.path.basename(row[0]))
    urllib.request.urlretrieve (row[0], os.path.join(output, os.path.basename(row[0])))
```
